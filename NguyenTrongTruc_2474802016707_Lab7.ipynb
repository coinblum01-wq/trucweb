## PHẦN 3: PHÂN TÍCH TẦN SỐ CỦA CÁC TỪ

```
fd = nltk.FreqDist(macbeth)
print("10 từ phổ biến nhất (chưa lọc gì):")
print(fd.most_common(10))
```

```
sw = set(nltk.corpus.stopwords.words('english'))
print("Số lượng stopword tiếng Anh:", len(sw))
print("10 stopword ví dụ:", list(sw)[:10])

macbeth_filtered = [w for w in macbeth if w.lower() not in sw]
print("Số từ còn lại sau khi loại stopword:", len(macbeth_filtered))

fd2 = nltk.FreqDist(macbeth_filtered)
print("10 từ phổ biến nhất (đã loại stopword, còn dấu câu):")
print(fd2.most_common(10))
```

```
punctuation = set(string.punctuation)
macbeth_filtered2 = [
    w.lower() for w in macbeth
    if w.lower() not in sw and w.lower() not in punctuation
]
fd3 = nltk.FreqDist(macbeth_filtered2)
print("10 từ phổ biến nhất (đã loại stopword và dấu câu):")
print(fd3.most_common(10))
```

## PHẦN 8: PHÂN TÍCH CẢM XÚC NGƯỜI DÙNG
Dùng thuật toán Naive Bayes để phân loại cảm xúc (positive/negative) dựa trên bộ dữ liệu `movie_reviews`.

```
reviews = nltk.corpus.movie_reviews
documents = [
    (list(reviews.words(fileid)), category)
    for category in reviews.categories()
    for fileid in reviews.fileids(category)
]
random.shuffle(documents)

first_review = ' '.join(documents[0][0])
print(first_review[:300], "...")
print("Nhãn:", documents[0][1])
```

```
def document_features(document, word_features):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['{}'.format(word)] = (word in document_words)
    return features

# Lưu ý: dùng toàn bộ từ vựng (~39.000 từ) sẽ chạy khá lâu.
# Ở đây giới hạn 2000 từ phổ biến nhất để chạy nhanh hơn trên Colab.
all_words = nltk.FreqDist(w.lower() for w in reviews.words())
word_features = list(all_words)[:2000]

featuresets = [(document_features(d, word_features), c) for (d, c) in documents]
print("Tổng số review:", len(featuresets))
```

```
train_set, test_set = featuresets[200:], featuresets[:200]
classifier = nltk.NaiveBayesClassifier.train(train_set)
acc = nltk.classify.accuracy(classifier, test_set)
print("Độ chính xác:", acc)
```

```
classifier.show_most_informative_features(10)
```

**Bài 6.** **Tìm định nghĩa và ví dụ của một từ cho trước bằng WordNet.**

```
from nltk.corpus import wordnet as wn

tu = 'computer'
for syn in wn.synsets(tu)[:3]:
    print(f"- {syn.name()}: {syn.definition()}")
    if syn.examples():
        print("    Ví dụ:", syn.examples())
```

**Bài 7.** **Tìm tập hợp các từ đồng nghĩa và trái nghĩa của một từ cho trước.**

```
tu = 'good'
synonyms, antonyms = set(), set()
for syn in wn.synsets(tu):
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print("Đồng nghĩa:", sorted(synonyms))
print("Trái nghĩa:", sorted(antonyms))
```

**Bài 9.** **So sánh độ giống nhau (Wu-Palmer similarity) giữa hai danh từ cho trước.**

```
tu1, tu2 = 'car', 'automobile'
syn1 = wn.synsets(tu1, pos=wn.NOUN)
syn2 = wn.synsets(tu2, pos=wn.NOUN)
print(f"similarity({tu1}, {tu2}) =", syn1[0].wup_similarity(syn2[0]))
```

**Bài 10.** **So sánh độ giống nhau (Wu-Palmer similarity) giữa hai động từ cho trước.**

```
tu1, tu2 = 'run', 'walk'
syn1 = wn.synsets(tu1, pos=wn.VERB)
syn2 = wn.synsets(tu2, pos=wn.VERB)
print(f"similarity({tu1}, {tu2}) =", syn1[0].wup_similarity(syn2[0]))
```

**Bài 11.** **Đếm số lượng tên nam và nữ trong kho ngữ liệu `names`; in ra 10 tên đầu tiên của mỗi loại.**

```
from nltk.corpus import names

male_names = names.words('male.txt')
female_names = names.words('female.txt')

print("Số tên nam:", len(male_names))
print("Số tên nữ :", len(female_names))
print("10 tên nam đầu tiên:", male_names[:10])
print("10 tên nữ đầu tiên :", female_names[:10])
```

**Bài 12.** **In ra 15 kết hợp ngẫu nhiên (tên, nhãn giới tính) từ kho `names`.**

```
labeled_names = (
    [(name, 'male') for name in male_names] +
    [(name, 'female') for name in female_names]
)
random.shuffle(labeled_names)

for item in labeled_names[:15]:
    print(item)
```
