***Cell 1: Khai báo thư viện***

```
import pandas as pd
import numpy as np
```

***Cell 2: Đọc dữ liệu và hiển thị 10 dòng đầu tiên***

```
# Sửa lại đường dẫn thêm chữ 'DuLieu/' ở phía trước
df = pd.read_csv('DuLieu/processed_dulieuxettuyendaihoc.csv', index_col='STT')
df.head(10)
```

***Cell 3: Xử lý dữ liệu khuyết thiếu (Missing value) cho cột DT***

```
# Điền các giá trị trống ở cột 'DT' bằng số 0
df['DT'] = df['DT'].fillna(0)
df
```

***Cell 4: Kiểm tra và xử lý dữ liệu khuyết thiếu cho cột T1***

```
# Kiểm tra số lượng giá trị trống ban đầu
print("Số lượng giá trị trống trước khi xử lý:", df['T1'].isnull().sum())
print("\nThống kê số lượng các giá trị:")
print(df['T1'].value_counts(dropna=False))
print("\nTỷ lệ các giá trị:")
print(df['T1'].value_counts(normalize=True, dropna=False))

# Thay thế các giá trị trống bằng giá trị trung bình của cột T1
df['T1'] = df['T1'].fillna(df['T1'].mean())

# Kiểm tra lại sau khi điền
print("\nSố lượng giá trị trống sau khi xử lý:", df['T1'].isnull().sum())
```

***Cell 5: Tính điểm trung bình môn (TBM) cho từng học kỳ***

```
# Tính toán điểm TBM cho các học kỳ 1, 2, 3 dựa trên công thức nhân hệ số
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
df['TBM3'] = (df['T3']*2 + df['L3'] + df['H3'] + df['S3'] + df['V3']*2 + df['X3'] + df['D3'] + df['N3']) / 10

# Hiển thị 5 dòng đầu của các cột TBM để kiểm tra
df[['TBM1', 'TBM2', 'TBM3']].head()
```

***Cell 6: Xếp loại học lực dựa trên điểm TBM***

```
def xeploai(diem):
    if diem < 5:
        return 'Y'
    elif diem < 6.5:
        return 'TB'
    elif diem < 8:
        return 'K'
    elif diem < 9:
        return 'G'
    else:
        return 'XS'

# Áp dụng hàm xếp loại cho cả 3 học kỳ
df['XL1'] = df['TBM1'].apply(xeploai)
df['XL2'] = df['TBM2'].apply(xeploai)
df['XL3'] = df['TBM3'].apply(xeploai)

# Hiển thị kết quả so sánh điểm và xếp loại
df[['TBM1', 'XL1', 'TBM2', 'XL2', 'TBM3', 'XL3']].head()
```

***Cell 7: Quy đổi điểm TBM sang thang điểm US (Thang điểm 4)***

```
# Quy đổi điểm từ thang 10 sang thang 4 bằng cách chia 10 rồi nhân 4
df['US_TBM1'] = df['TBM1'] / 10 * 4
df['US_TBM2'] = df['TBM2'] / 10 * 4
df['US_TBM3'] = df['TBM3'] / 10 * 4

# Hiển thị kết quả quy đổi
df[['TBM1', 'US_TBM1', 'TBM2', 'US_TBM2', 'TBM3', 'US_TBM3']].head()
```

***Cell 8: Tính toán kết quả xét tuyển đại học (KQXT) dựa theo khối thi***

```
def tinh_kqxt(row):
    khoi = row['KT']
    # Nếu thuộc khối A hoặc A1 thì môn thứ nhất nhân hệ số 2
    if khoi in ['A', 'A1']:
        diem = (row['DH1']*2 + row['DH2'] + row['DH3']) / 4
    # Nếu thuộc khối B thì môn thứ hai nhân hệ số 2
    elif khoi == 'B':
        diem = (row['DH1'] + row['DH2']*2 + row['DH3']) / 4
    # Các khối còn lại tính trung bình cộng thông thường
    else:
        diem = (row['DH1'] + row['DH2'] + row['DH3']) / 3

    # Điểm từ 5 trở lên thì Đạt (1), ngược lại là Rớt (0)
    return 1 if diem >= 5 else 0

# Áp dụng hàm tính KQXT cho từng dòng (axis=1)
df['KQXT'] = df.apply(tinh_kqxt, axis=1)

# Hiển thị kết quả xét tuyển cùng các điểm thành phần
df[['KT', 'DH1', 'DH2', 'DH3', 'KQXT']].head()
```
